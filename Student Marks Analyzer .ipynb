```python
import numpy as np
import matplotlib.pyplot as plt

numbers = np.array([
    [1, 78, 67, 90],
    [2, 34, 67, 89],
    [3, 87, 56, 87],
    [4, 69, 87, 45],
    [5, 56, 57, 95]
])

print("note: the marks is out of 100")

print("\n Total marks obtained by student 1: ", np.sum(numbers[0][1:]))
print("\n Total marks obtained by student 2: ", np.sum(numbers[1][1:]))
print("\n Total marks obtained by student 3: ", np.sum(numbers[2][1:]))
print("\n Total marks obtained by student 4: ", np.sum(numbers[3][1:]))
print("\n Total marks obtained by student 5: ", np.sum(numbers[4][1:]))

total_mathes = np.sum(numbers[:, 1])
total_physics = np.sum(numbers[:, 2])
total_english = np.sum(numbers[:, 3])

print("\n Total mathes number obtained by all student: ", total_mathes/5)
print("\n Total physics number obtained by all student: ", total_physics/5)
print("\n Total english number obtained by all student: ", total_english/5)

print("\n Percentage of student 1: ", np.sum(numbers[0][1:])/3)
print("\n Percentage of student 2: ", np.sum(numbers[1][1:])/3)
print("\n Percentage of student 3: ", np.sum(numbers[2][1:])/3)
print("\n Percentage of student 4: ", np.sum(numbers[3][1:])/3)
print("\n Percentage of student 5: ", np.sum(numbers[4][1:])/3)

x = numbers[:, 0]
y = [np.sum(row[1:])for row in numbers]


plt.bar(x, y, color = 'red')

plt.title("total number of student")
plt.xlabel("roll number")
plt.ylabel("total numbers")
plt.ylim(0, 300)
plt.grid(linestyle = '--', linewidth = 2, alpha = 0.3)

plt.show()
```
