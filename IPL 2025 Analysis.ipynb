```python
 # --------------------------------------  Phase - 1  ------------------------------------------

import pandas as pd # importing pandas for data analysis
import matplotlib.pyplot as plt # importing matplotlib for data visualisation
import numpy as np # importing numpy for performing numerical operations


# Loading the data:
data = pd.read_csv("IPL2025Batters.csv")

# Summary of data:

print("\n Shape of the DATA: \n", data.shape) # state the shape of the data

print("\n description of the DATA: \n", data.describe()) # Numerical information 

print("\n Statestical summary of the DATA: \n")
data.info()  # Shows types + non-null values

print("\n first 5 rows of the DATA: \n", data.head()) # first 5 rows

print("\n last 5 rows of the DATA: \n", data.tail()) # last 5 rows

print("\n rows of the DATA: \n", data.shape[0]) # List no. of rows

print("\n column of the DATA: \n", list(data.columns)) # Name all columns

print("\n Null values in each columns \n", data.isnull().sum()) # Total count of null values in each columns

data = data.dropna() # Drop null rows
data = data.drop_duplicates() # Drop duplicate rows

# Remove "*" from 'HS' and convert to int
data['HS'] = data['HS'].astype(str).str.replace('*', '', regex = False)
data['HS'] = pd.to_numeric(data['HS'], errors = 'coerce')

# Convert 'AVG' to numeric DTYPE
data['AVG'] = pd.to_numeric(data['AVG'], errors = 'coerce')

# Verify the final structure:

print("\n Data type of data: \n", data.dtypes)
print("\n shape of data: \n", data.shape)

# --------------------------------------  Phase - 2  ------------------------------------------

# by using sort function sort the Top 10 run scrorers in the dataset:

top10_runs = data.sort_values(by = 'Runs', ascending = False).head(10)
print("\n Top 10 run scrores: ", top10_runs)

# Data visualization section by Matplotlib:

plt.figure(figsize = (10, 6))
plt.barh(top10_runs['Player Name'], top10_runs['Runs'], color = 'skyblue', label = 'Top 10 run scrorers')
plt.title('Top 10 player with highest runs')
plt.xlabel('Runs of the player')
plt.ylabel('Name of the player')
plt.legend()
plt.grid(True, color = 'yellow', linestyle = '--', alpha = 0.5)
plt.gca().invert_yaxis()
plt.show()

# by using sort function we sort top 10 player with highest strike rate:

data = data[data['Inn'] >= 5]
top10_sr = data.sort_values(by = 'SR', ascending = False).head(10)
print(top10_sr)

# Data visualisation section:

plt.figure(figsize = (10, 6))
plt.barh(top10_sr['Player Name'], top10_sr['SR'], color = 'orange', label = 'Top 10 player by SR')
plt.title('top 10 player with highest strike rate')
plt.xlabel('Strike Rate')
plt.ylabel('Player name')
plt.legend()
plt.grid(color = 'green', alpha = 0.5, linestyle = '--')
plt.gca().invert_yaxis()
plt.show()

# by using sort function sort the Top 10 players by 6s

top10_6s = data.sort_values(by = '6s', ascending = False).head(10)
print(top10_6s)

# Data visualization section:

plt.figure(figsize = (10, 6))
plt.barh(top10_6s['Player Name'], top10_6s['6s'], color = 'red', label = 'Top 10 player by 6s')
plt.title('Top 10 player with highest 6s')
plt.xlabel('6s of the player')
plt.ylabel('Name of the player')
plt.legend()
plt.grid(True, color = 'blue', linestyle = '--', alpha = 0.5)
plt.gca().invert_yaxis()
plt.show()

# by using sort function sort the Top 10 players by 4s

top10_4s = data.sort_values(by = '4s', ascending = False).head(10)
print(top10_4s)

# Data visualization section:

plt.figure(figsize = (10, 6))
plt.barh(top10_4s['Player Name'], top10_4s['4s'], color = 'pink', label = 'Top 10 player by 4s')
plt.title('Top 10 player with highest 4s')
plt.xlabel('4s of the player')
plt.ylabel('Name of the player')
plt.legend()
plt.grid(True, color = 'yellow', linestyle = '--', alpha = 0.5)
plt.gca().invert_yaxis()
plt.show()


# by using sort function sort the Top 10 players by their AVG scores:

top10_avg = data.sort_values(by = 'AVG', ascending = False).head(10)
print(top10_avg)

# Data visualization section:

plt.figure(figsize = (10, 6))
plt.barh(top10_avg['Player Name'], top10_avg['AVG'], color = 'purple', label = 'Top 10 player by AVG')
plt.title('Top 10 player with highest AVG')
plt.xlabel('AVG of the player')
plt.ylabel('Name of the player')
plt.legend()
plt.grid(True, color = 'skyblue', linestyle = '--', alpha = 0.5)
plt.gca().invert_yaxis()
plt.show()

# Creating a new columns Boundries:

data['Boundries'] = data['4s'] + data['6s']

# by using sort function sort the Top 10 players by Boundries

top10_b = data.sort_values(by = 'Boundries', ascending = False).head(10)
print(top10_b)

# Data visualization section:

plt.figure(figsize = (10, 6))
plt.barh(top10_b['Player Name'], top10_b['Boundries'], color = 'green', label = 'Top 10 player by Boundries')
plt.title('Top 10 player with highest Boundries')
plt.xlabel('Boundries of the player')
plt.ylabel('Name of the player')
plt.legend()
plt.grid(True, color = 'yellow', linestyle = '--', alpha = 0.5)
plt.gca().invert_yaxis()
plt.show()

# some numerical operation by numpy:

high6s = np.max(data['6s'])
high4s = np.max(data['4s'])
highb = np.max(data['Boundries'])
low4s = np.min(data['4s'])
low6s = np.min(data['6s'])
lowb = np.min(data['Boundries'])
avgsr  = np.mean(data['SR'])
medsr = np.median(data['SR'])

print("\n Maximum sixes of a individual: \n", high6s)
print("\n Maximum fours of a individual: \n", high4s)
print("\n Maximum Boundries of a individual: \n", highb)
print("\n Minimum Boundries of a individual: \n", lowb)
print("\n Mainimum fours of a individual: \n", low4s)
print("\n Mainimum sixes of a individual: \n", low6s)
print("\n average strike rate of all palyers: \n", avgsr)
print("\n midian strike rate of all palyers: \n", medsr)
```
